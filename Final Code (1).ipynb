```python
# Reading the Response excel sheet i.e. input 1

xlsx = pd.ExcelFile(r'C:\Users\Shweta Kapoor\Desktop\project ss jupiter\Form Response.xlsx')
rawdata = pd.read_excel(xlsx)
rawdata = rawdata.iloc[1:,1:30]
rawdata
```

```python
# Reading the profiles excel sheet i.e. input 2
profile = pd.read_excel(r'C:\Users\Shweta Kapoor\Desktop\project ss jupiter\Profiles.xlsx')
profile = profile.iloc[2:,:]
profile
```

```python
traits=list(profile.columns)
traits[0]='Email'
jobs = list(profile['Profiles'])
jobs.insert(0,'Email')
```

```python
print(jobs)
```

```python
# Now we are going to calculate the score for each candidate
```

```python
# list negative questions
n=[5,7,12,14,16,17,18,21,25,27]
score=0
l=[]
lst=[]
for index, row in rawdata.iterrows() :
    ind=1
    l.append(row[0])
    for k in range(1,8):
        ind=k
        for j in range(0,4):
            if ind not in n :
                score=score+row[ind]
            else:
                score=score+(row[ind]*-1)+6
            ind=ind+7;
        l.append(score)
        score=0
    lst.append(l)
    l=[]
```

```python
score.head()
```

```python
# making a dataframe for our final scores
score = pd.DataFrame(lst, columns = traits)
score.head()
```

```python
dst=[]
ed=0
nor=52.91502622129181
eucledian=[]
for index, row in score.iterrows():
    #if index == 1:
        #break;
    dst.append(row[0])
    for i,j in profile.iterrows():
        d=0
        for k in range (1,8):
            d=d+((row[k]-j[k])**2)
        ed=math.sqrt(d)
        ed=ed/nor
        dst.append(ed)
    eucledian.append(dst)
    dst=[]
```

```python
distance = pd.DataFrame(eucledian, columns=jobs)
temp = distance.iloc[:,1:]
```

```python
dclass=[]
for index,row in temp.iterrows():
    l=row.tolist()
    x=l.index(min(l))
    col=temp.columns[x]
    dclass.append(col)
```

```python
distance['Class'] = dclass
score['Class'] = dclass
score
```

```python
writer = pd.ExcelWriter(r'C:\Users\Shweta Kapoor\Desktop\project ss jupiter\Score.xlsx', engine='openpyxl') 

#Add dataframe to excel file 
score.to_excel(writer,index=False)

writer.save()
```
